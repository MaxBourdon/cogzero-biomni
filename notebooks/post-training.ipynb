```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel

base_model_name = "/mlbio_scratch/kodryan/hf_downloads/Qwen3-32B/"
lora_path = "/mlbio_scratch/kodryan/hf_downloads/Qwen3-32B-Biomni-LoRA/"

# 1) Load base model
model = AutoModelForCausalLM.from_pretrained(
    base_model_name,
    torch_dtype="auto",
    device_map="auto",
)

# 2) Attach LoRA adapter
model = PeftModel.from_pretrained(model, lora_path)
```

```python
# 3) Merge LoRA into the base weights and drop PEFT wrappers
model = model.merge_and_unload()

# 4) (Optional) save as a plain Transformers model
save_path = "/mlbio_scratch/kodryan/hf_downloads/Qwen3-32B-Solver-Biomni"
model.save_pretrained(save_path)
tokenizer = AutoTokenizer.from_pretrained(base_model_name)
tokenizer.save_pretrained(save_path)
```
